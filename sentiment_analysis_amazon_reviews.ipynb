**Step 2:** Establish the data cleaning process.
* **Explanation & Justification:** It can be observed that there are numbers, years indicating a date, unnecessary text explanations within brackets , html tags, unwanted whitespace, and other unwanted wierd characters within the text that are not useful, i.e., they wouldn't inform the model whether the review is of a positive or negative sentiment. Since we are only interested in the text, we want to remove parts of the text that is not useful. <br> (Note, the reason why we want to ignore text within brackets is because these are usually explanations for abstract topics to give a better understanding to the reader about the review, it doesn't indicate whether the review is of a positive or negative sentiment)<br>

**Step 3:** Use regular expressions to clean the text according to **Step 2**.

```python
import re

# define a function to clean a text
# I used regex101.com to test my regular expressions before use.
def clean_text(text):
    
    # remove hyper-links
    text = re.sub(r"https:[\S]+", " ", text)
    
    # remove unnecessary text explanations within brackets
    text = re.sub(r"\(([^)]+)", " ", text)
    
    # remove html tags? not sure why there are html tags in some of the text
    # but removing them nonetheless
    text = re.sub(r"\<([^>]+)", " ", text)
    
    # remove any numbers or fractions or brackets or years representation of dates
    text = re.sub(r"[(\d\/)]+ | [\d]+\S", " ", text)

    # remove new line characters (\n)
    text = re.sub(r"\\n"," ", text)
    
    # remove any weird characters
    text = re.sub(r"[^A-Za-z']+", " ", text)
        
    # remove unnecessary whitespace
    text = re.sub(r"\s{2,}", " ", text)
        
    # convert text to lowercase
    text = text.lower().strip()
    
    return text
```

```python
# clean the text within the dataset using the clean_text(text) function
amazon_reviews["Text"] = amazon_reviews["Text"].apply(clean_text)
```

## 3.1 List of Steps to Build the Model

* **Step 1:** Split the sampled data into train and test set. <br>
* **Step 2:** Represent the dataset text reviews as a numeric vector using Bag of Words (BOW) to extract text features. <br>
* **Step 3:** Build and train the Logistic Regression model (that uses a maximum likelihood learning algorithm) using the data features in training set, then use the model to make predictions on the sentiment for the data in test set based on the test data features.

## 3.2 Train-Test Split

We need to first split the data into train and test set. Train set data will be used to develop a text classification model and test set data will be used to evaluate the model later on.

```python
import numpy as np
from sklearn.model_selection import train_test_split

# define random state to make random processes reproducible
random_state = np.random.RandomState(0)
```

```python
# Define X(features) and y(target)
X = amazon_reviews["Text"]
y = amazon_reviews["Label"]

# 70% of the reviews should be used for training and the remaining 30% for the testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=random_state)
```

## 3.3 Represent Text as Bag of Words (BOW) 

Models like Logistic Regression can not directly handle text data. They only accept numeric data. Therefore, text should be converted to a numeric vector before inputting to the model. Bag of Words (BOW) is one of the most commonly used model to represent text as numeric vector.

```python
# CountVectorizer is a great tool provided by the scikit-learn library in Python. It is used to transform a 
# given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text.
from sklearn.feature_extraction.text import CountVectorizer
!pip install nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

The text can contain certain stopwords such as 'is', 'the', 'are', etc. which are not useful for identifying sentiments. Other common words found will also be removed.

```python
# define stop words and also add custom stop words such as 'product'
stop_words = stopwords.words('english')
# stop_words.extend(['food', 'product', 'foods', 'amazon']) ...to extend?
# stop_words.extend(['excited', 'guess', 'away', 'stick'])

# define CountVectorizer object
# min_df=0.01 means words that belong to only 1% of the all texts will be excluded in the BOW
vectorizer = CountVectorizer(stop_words=stopwords.words('english'), min_df=0.01)

# fit the vectorizer object to train data
vectorizer.fit(X_train)

# get the BOW for train data
X_train_BOW = vectorizer.transform(X_train)
```

```python
# visualize BOW
training_data_BOW = pd.DataFrame(data=X_train_BOW.toarray(), columns=vectorizer.get_feature_names_out()).T

# display the first 5 rows of the dataframe
training_data_BOW.head(5)
```

In the above dataframe, the row labels are list of words extracted from the reviews and column labels are text review numbers for the training data. The value inside the dataframe indicates number of occurrences of a particular word in the given review. For example, the word 'absolutely' occurs thrice in the review 19629.

## 3.3 Build the Logistic Regression Model & Make Predictions

Logistic regression belongs to a family of linear models that can be used to predict categorical outcome such as positive or negative sentiment, spam or not spam email, etc. In this case, it will take the bag of words as input and based on that it will predict whether the text review has 'pos' or 'neg' sentiment.

```python
from sklearn.linear_model import LogisticRegression
```

```python
# define lgositic regression model object
log_reg = LogisticRegression(random_state=random_state)

# fit the model to training data
log_reg.fit(X_train_BOW, y_train)
```

```python
# make prediction on test data

# create BOW for test data
X_test_BOW = vectorizer.transform(X_test)

# predict using log_reg object
y_test_log_reg_predictions = log_reg.predict(X_test_BOW)

# view the predictions
y_test_log_reg_predictions
```

### Precision & Recall (Negative Class)

For the negative 'neg' class, 
* **Precision = TN/(TN+FN)**, where TN is True Negative and FN is False Negative
* **Recall = TN/(TN+FP)**, where TN is True Negative and FP is False Postive

Since the baseline model assigns positive sentiment label to all test samples, it will never identify a text review that is of negative sentiment to be of a negative sentiment. Therefore **TN** for the baseline model will be **0** for all test samples. 

* The baseline model correctly identifies 0 negative sentiment reviews as negative (True Negative - TN)
* The baseline model incorrectly identifies 0 positive sentiment reviews as negative (False Negative - FN)
* The baseline model incorrectly identifies some negative sentiment reviews as positive (False Positve - FP)

This implies that precision and recall for the baseline model will be 0 based on the formulas provided above. <br> 

(Technically precision would be undefined, since we would end up dividing 0 by 0, as FN would be 0 due to the baseline model never identifying negative sentiments for reviews that have a positive sentiment. But recall would be actually 0 since we divide TN, which is 0, by FP, and FP is greater than 0 since there would be some text reviews of negative sentiment that the baseline model would identify as having a positive sentiment.)

* Precision = TN/(TN + FN) = 0/(0 + 0) = undefined, due to division by 0
* Recall = TN/(TN + FP) = 0/(0 + x) = 0, where x > 0

# 6.2 Which words classify a review as positive or negative?

Following function will identify top words which were marked important by the logistic regression model in making predictions.

```python
# function adapted from https://aneesha.medium.com/visualising-top-features-in-linear-svm-with-scikit-learn-and-matplotlib-3454ab18a14d
def plot_coefficients(classifier, feature_names, top_features=10):
    coef = classifier.coef_.ravel()
    top_positive_coefficients = np.argsort(coef)[-top_features:]
    top_negative_coefficients = np.argsort(coef)[:top_features]
    top_coefficients = np.hstack([top_negative_coefficients, top_positive_coefficients])
    # create plot
    plt.figure(figsize=(15, 5))
    colors = ['red' if c < 0 else 'blue' for c in coef[top_coefficients]]
    plt.bar(np.arange(2 * top_features), coef[top_coefficients], color=colors)
    feature_names = np.array(feature_names)
    plt.xticks(np.arange(0, 2 * top_features), feature_names[top_coefficients], rotation=90, ha='right')
    plt.xlabel("Important Words")
    plt.ylabel("Model Coefficient")
    plt.title("Important words with their model coefficient")
    plt.show()
```

```python
plot_coefficients(log_reg, vectorizer.get_feature_names_out())
```

The plot shows 20 words which played a significant role in the predictions made by the logistic regression model. The red bars show the top 10 words with negative model coefficient and the blue bars show the top 10 words with positive model coefficient values.

We observe that the words 'disappointed', 'waste', 'unfortunately', 'excited', 'bad', 'guess', 'away', 'stick', 'please', and 'money', played a significant role in predicting that the review is of a negative sentiment. We also observe that the words 'awesome', 'excellent', 'refreshing', 'perfect', 'thank', 'delicious', 'fantastic', 'loves', 'smooth', and 'yummy' played a significant role in predicting that the review is of a positive sentiment.

# Task 7: About Classification with Unbalanced Data...

* Classification with unbalanced data...
* What is the issue with unbalanced data? How can we handle it in a better way? 

The data is unbalanced when it has has a certain "amount" of records of one class in abundance in comparison to other classes. Now how much this "amount" needs to be is subjective for the dataset to be considered "unbalanced". For example, in our reviews dataset, we had much more positive sentiment reviews (83%, the majority class) in comparison to negative sentiment reviews (17%, the minority class). There is a disparity in the classes and hence the dataset would be considered as "unbalanced". <br> <br>

The issue with unbalanced data is that training a model/classifier on the data would result in a biased model that tends to predict the majority class, since the model frequently observes more records and features of the majority class (also, metrics like accuracy woudn't be reliable in evaluating such a model due to the class imbalances). <br>
There are multiple ways to resolve the issues posed by unbalanced data, but one way is to sample an arbitrary portion of the data that has an equal number of records of all classes present in the dataset (this is called under-sampling). Using this sampled data will ensure that the data is balanced (and more complete), which results in improved data quality.
